```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib import patches

print("TensorFlow version:", tf.__version__)
```

```python
# TF Hub model hỗ trợ detection + segmentation
MODEL_URL = "https://tfhub.dev/google/faster_rcnn/openimages_v4/inception_resnet_v2/1"
detector = hub.load(MODEL_URL)
print("Model loaded!")
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("sample.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,8))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
# Chuyển ảnh sang float32 và chuẩn hóa về [0,1]
input_tensor = tf.convert_to_tensor(img_rgb, dtype=tf.float32) / 255.0
input_tensor = tf.expand_dims(input_tensor, 0)

# Dùng signature để detect
detector_fn = detector.signatures["default"]
result = detector_fn(input_tensor)
print("Detection complete!")
```

```python
import numpy as np

# Chuẩn hóa dữ liệu
boxes = result["detection_boxes"][0].numpy()
scores = result["detection_scores"][0].numpy()
classes = result["detection_class_entities"][0].numpy()

# Nếu scores là float, chuyển sang array 1 phần tử
if np.isscalar(scores):
    scores = np.array([scores])
    boxes = np.expand_dims(boxes, 0)
    classes = np.array([classes])

plt.figure(figsize=(12,12))
plt.imshow(img_rgb)
height, width, _ = img_rgb.shape

for i in range(len(scores)):
    if scores[i] > 0.5:
        y1, x1, y2, x2 = boxes[i]
        rect = patches.Rectangle((x1*width, y1*height),
                                 (x2-x1)*width, (y2-y1)*height,
                                 linewidth=2, edgecolor='r', facecolor='none')
        plt.gca().add_patch(rect)
        plt.text(x1*width, y1*height-5, classes[i].decode("utf-8"), color='yellow', fontsize=12)

plt.axis('off')
plt.show()
```

```python
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
import cv2
import tensorflow as tf

# Upload file
uploaded = files.upload()

# Kiểm tra xem có file nào được chọn không
if len(uploaded) == 0:
    print("Bạn chưa chọn file nào để upload. Vui lòng chọn ảnh để chạy instance segmentation.")
else:
    for fname in uploaded.keys():
        print(f"Đang xử lý ảnh: {fname}")
        # Đọc ảnh
        img = cv2.imread(fname)
        if img is None:
            print(f"Không thể đọc ảnh: {fname}")
            continue
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Chuyển ảnh sang float32 và chuẩn hóa [0,1]
        input_tensor = tf.convert_to_tensor(img_rgb, dtype=tf.float32) / 255.0
        input_tensor = tf.expand_dims(input_tensor, 0)

        # Dùng TF Hub signature để detect
        detector_fn = detector.signatures["default"]
        result = detector_fn(input_tensor)

        # Lấy bounding box + score + class
        boxes = result["detection_boxes"][0].numpy()
        scores = np.squeeze(result["detection_scores"][0].numpy())
        classes = np.squeeze(result["detection_class_entities"][0].numpy())

        # Nếu chỉ có 1 detection, ép thành array 1D/2D
        if np.isscalar(scores):
            scores = np.array([scores])
            boxes = np.expand_dims(boxes, 0)
            classes = np.array([classes])

        # Vẽ kết quả
        plt.figure(figsize=(12,12))
        plt.imshow(img_rgb)
        height, width, _ = img_rgb.shape

        for i in range(len(scores)):
            if scores[i] > 0.5:  # chỉ hiển thị confident > 0.5
                y1, x1, y2, x2 = boxes[i]
                rect = patches.Rectangle(
                    (x1*width, y1*height),
                    (x2-x1)*width,
                    (y2-y1)*height,
                    linewidth=2, edgecolor='r', facecolor='none'
                )
                plt.gca().add_patch(rect)
                plt.text(x1*width, y1*height-5,
                         classes[i].decode("utf-8"),
                         color='yellow', fontsize=12)

        plt.axis('off')
        plt.show()
```
